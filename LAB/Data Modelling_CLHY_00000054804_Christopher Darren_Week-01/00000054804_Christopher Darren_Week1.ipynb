```python
import pandas as pd
```

```python
#membaca data dari file yang tersedia
data = pd.read_csv(r'D:\SEMESTER 4\IS411 Data Modelling\LAB\MBA_Python.csv', header=0, delimiter=';')

data.info()
```

```python
data.dropna()
```

```python
data['Quantity'].min()

dataPositif = data[data['Quantity'] >= 0]
dataPositif['Quantity'].min()
```

```python
basket = (dataPositif.groupby(['Invoice','Description'])['Quantity'].sum().unstack()
         .reset_index().fillna(0).set_index(['Invoice']))
basket
```

```python
def encode(qty):
    if qty >= 1:
        return 1
    if qty <= 0:
        return 0

basketSimplified = basket.applymap(encode)
basketSimplified
```

```python
finalBasket = basketSimplified[(basketSimplified > 0).sum(axis=1) >= 2]
```

```python
from mlxtend.frequent_patterns import apriori

frequent_item = apriori(finalBasket, min_support=0.03, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_item
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_item, metric='lift', min_threshold=1).sort_values('lift',ascending=False).reset_index(drop=True)
```
