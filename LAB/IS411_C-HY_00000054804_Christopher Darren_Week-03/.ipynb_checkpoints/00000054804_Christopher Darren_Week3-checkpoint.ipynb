## Data Mining Methodology

```python
#import data

import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv(r"D:\SEMESTER 4\IS411 Data Modelling\LAB\Bahan Modul 3\MBA_latihan.csv", delimiter=';')
dataset.head(10)
```

```python
dataset.info()
```

```python
#menampilkan informasi untuk mengetahui kolum/atribut data yang memiliki data null (empty) dalam barisnya

print(dataset.isnull().sum())
```

```python
#menghapus kolom yang memiliki data null

new_dataset = dataset.dropna(axis = 1)
new_dataset.info()
```

```python
#menghapus baris yang memiliki data null

new_dataset2 = dataset.dropna(axis = 0)
new_dataset2.info()
```

```python
#mengisi data null dengan nilai 0

default_dataset = dataset.fillna(0)
default_dataset.info()
```

```python
#mengisi bagian null pada kolom 'Description' dengan 'no description'

default_dataset_spec = dataset.copy()
default_dataset_spec['Description'] = default_dataset_spec['Description'].fillna('no description')
```

```python
#mengisi bagian null pada kolom 'Customer ID' dengan 0

default_dataset_spec['Customer ID'] = default_dataset_spec['Customer ID'].fillna(0)
```

```python
#menampilkan informasi data

default_dataset_spec.info()
```

```python
#membuat kolom baru bernama 'mean_dataset'

mean_dataset = dataset.copy()
mean_dataset['mean_quantity'] = np.nan
mean_dataset.info()
```

```python
#mengisi kolom 'mean_quantity' dengan nilai rata-rata (mean) dari kolom 'Quantity'

mean_dataset['mean_quantity'] = mean_dataset['mean_quantity'].fillna(mean_dataset['Quantity'].mean())
mean_dataset.info()
mean_dataset
```

```python
data_cond = dataset.copy()

data_cond['Description'] = np.where(np.logical_and(data_cond['Description'].isnull(), data_cond['Quantity'] > 10), 'bulk',
                                   np.where(data_cond['Description'].isnull(), 'not clear', data_cond['Description']))
data_cond.info()
```

```python
#memastikan data sudah terisi sesuai kondisi yang ditentukan

data_cond[data_cond['Description'] == 'bulk']
```

```python
data_cond[data_cond['Description'] == 'not clear']
```

### Containts vs Match

```python
data_contain = dataset['Description'].str.contains('LIGHTS')
print(data_contain)
```

```python
data_match = dataset['Description'].str.match('LIGHTS')
print(data_match)
```

```python
data_match2 = dataset['Description'].str.match('FELTCRAFT PRINCESS LOLA DOLL')
print(data_match2)
```

##### perbedaan output antara contains() dan match()
#### kalau contains itu ada berarti baris yang ditentukan harus berisi, di suatu tempat di dalamnya, string yang ditentukan, intinya yang ada hubungannya bisa dimasukkan pada function ini cth, ada kata "LIGHTS" kemudian ada 1 record yang isinya "LIGHTS BLABLAL" maka akan tetap diinput.

#### kalau match itu berarti baris yang ditentukan harus cocok dengan string yang ditentukan, contoh pada "FELTCRAFT PRINCESS LOLA DOLL"


```python
data_replace = dataset['Description'].str.replace('LIGHTS','LAMP')
data_replace
```
